```python
import numpy as np 

print("Welcome To Noni Calucalator Website")
print("\n")
print("Select the equation to solve:")
print("1. Cubic Equation")
print("2. Quadratic Equation")
print("3. Quartic Equation")
print("4. Exit")
lapp = int(input("Enter your option"))

match lapp:
    case 1:
        print("Cubic Equation")
        a = float(input("Input a"))
        b = float(input("Input b"))
        c = float(input("Input c"))
        d = float(input("Input d"))

        coeff = [a, b, c, d]
        roots = np.roots(coeff)

        print("Roots are: ",roots)
        
    case 2:
        print("Qudratic Equation")
        e = float(input("Input a"))
        f = float(input("Input b"))
        g = float(input("Input c"))

        eff = [e, f, g]
        roots = np.roots(eff)
    
        print("Roots are: ", roots)
    case 3:
        print("Quartic Equation")
        z = int(input("Input a"))
        y = int(input("Input b"))
        x = int(input("Input c"))
        w = int(input("Input d"))
        v = int(input("Input e"))

        hih = [z, y, x, w, v]
        roots = np.roots(hih)

        print("Roots are: ",roots)

    case 4:
        print("Thank you")
```
