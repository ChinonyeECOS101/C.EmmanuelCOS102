```python
import pandas as pd
import tkinter as tk
from PIL import Image, ImageTk
from tkinter import messagebox

# using pandas to import the csv file which contain the names and departments of the 
#staffs
df = pd.read_csv(r"C:\Users\turninpointKS\Desktop\cos_102\GIG-logistics_dep.csv")

# Creating a tkinter window
def welcomeMessage(username, department_list):
    window = tk.Toplevel(root)
    window.title("Name Verification")
    window.geometry("500x200")

    label_1 =tk.Label(window, text=f"Welcome, {username} You are an Employee\n",fg="brown", font=("Arial Rounded MT", 15, "bold italic" ))
    label_1.pack()
    label_2 = tk.Label(window, text="Other members of your department:", fg="skyblue", font=("Bahnschrift SemiBold", 17,
                                                                                             "underline italic bold"))
    label_2.pack()
    
    # Fortmat the list of other members of the user department
    department_str = "\n".join(department_list)
    tk.Label(window, text=department_str,fg="blue", font=("Arial", 15, "bold italic")).pack()

    #Run the tkinter event loop
    root.mainloop()
    
# Creating the submit button and it outputs
def login():    
    username = username_entry.get()
    department = department_entry.get()
    # Checking if the specific user and department match exists
    is_username = df[(df["Name"] == username) & (df["Department"] == department)]

    if not is_username.empty:
        # If they exist, get all OTHERS members of that same department
        # this filters the main df for the department, but excludes the current user
        others = df[(df["Department"] == department) & (df["Name"] != username)]
        department_list = others["Name"].tolist()
        welcomeMessage(username, department_list)
    else:
        # if there is no match print
        messagebox.showerror("ERROR", "Employee does not exist")

# Creating main window
root = tk.Tk()
root.title("Employee Checker")
root.geometry("500x200")

# Creating username label and entry
username_label = tk.Label(root, text="Employee's name:",fg="purple", font=("Times New Roman", 18, " bold"))
username_label.pack()
username_entry = tk.Entry(root, fg="green", font=("Bahnschrift SemiBold", 10, "italic bold"))
username_entry.pack()

# Create department label and entry
department_label = tk.Label(root, text="Department:",fg="purple", font=("Times New Roman", 18, " bold"))
department_label.pack()
department_entry = tk.Entry(root,fg="green", font=("Bahnschrift SemiBold", 10, "italic bold"))
department_entry.pack()

#creating the login button
login_button = tk.Button(root, text="Login", command=login)
login_button.pack()

# run the main loop
root.mainloop()

    
```
